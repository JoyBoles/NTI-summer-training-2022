# Load necessary libraries

# Basic graph

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.show()
```

```python
y=[2,6,8]

plt.plot(x,y)
plt.show()
```

```python
plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
#fontdict is a dictionary for specifying some font properties
plt.xlabel('X-axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y-axis')
plt.plot(x,y)
plt.show()
```

```python
x=[5,3,8,2,9]
y=[4,2,7,3,1]

plt.xticks([1,2,3,4,4.5,5,10,13])  #returns labels and their location on x-axis
plt.yticks([2,4,6,6.25,8,6,12,16]) #returns labels and their location on y-axis

plt.plot(x,y)

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x, y, label='2x', color='#AFBECD')
plt.xticks(np.arange(10))
plt.yticks(np.arange(10))

plt.legend(loc='lower right') #this makes the plot label appear (by default loc='upper right')

plt.show()
```

We can pass letter 'y', full word'yellow', or '#hexadecimal number' to color parameter in plot function

```python
plt.plot(x, y, label='2x', color='y', linewidth=3, marker='.', markersize=15, markeredgecolor='red', markerfacecolor='g')
plt.xticks(np.arange(10))
plt.yticks(np.arange(10))
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='2x', linestyle='--')
plt.xticks(np.arange(10))
plt.yticks(np.arange(10))
plt.legend()
plt.show()
```

fmt: str, optional
A format string, e.g. 'ro' for red circles. See the Notes section for a full description of the format strings.

Format strings are just an abbreviation for quickly setting basic line properties. All of these and more can also be controlled by keyword arguments.

This argument cannot be passed as keyword.

#### fmt is format in this order: [color][marker][line]
example: 'r^--' is [r][^][--]
color must be specified as a letter not a word or hexadecimal
* If we didn't specify the color, it wil be default blue
* If we didn't specify the marker, there will be no marker
* If we didn't specify the line, there will be no line

```python
plt.plot(x, y, 'r^--', label='2x')
plt.show()
```

```python
plt.plot(x, y, '^--', label='2x')
plt.show()
```

```python
plt.plot(x, y, 'r--', label='2x')
plt.show()
```

```python
plt.plot(x, y, '--', label='2x')
plt.show()
```

```python
plt.plot(x, y, '^', label='2x')
plt.show()
```

* If we didn't specify anything, color, marker, line will be default values i.e. blue, no marker, solid line

```python
x2=np.arange(0,5,0.25)
plt.plot(x2, 'g^-.', label='x')
plt.plot(x2,x2**2, 'm.--', label='x^2')
plt.xticks(np.arange(25))
plt.yticks(np.arange(25))
plt.legend()
plt.show()
```

```python
plt.plot(x2[:6], x2[:6]**2, 'r', label='x2^2')
plt.plot(x2[5:], x2[5:]**2, 'y--', label='X2^2')
plt.xticks(np.arange(25))
plt.yticks(np.arange(25))
plt.legend()
plt.show()
```

# Bar chart

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels, values)
plt.show()
```

```python
type(bars)
```

```python
bars=plt.bar(labels, values)
plt.figure(figsize=(9,9))
plt.show()
```

```python
bars=plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
plt.show()
```

```python
bars=plt.bar(labels, values)
patterns=['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```
