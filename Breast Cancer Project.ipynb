###### From pie chart and the count plot of the status, we need to perform resampling because data is extremely biased to alive

### Using LabelEncoder to transform categorical variables into continuous variables.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_race = LabelEncoder()
label_encoder_marital_status = LabelEncoder()
label_encoder_t_stage = LabelEncoder()
label_encoder_n_stage = LabelEncoder()
label_encoder_6th_stage = LabelEncoder()
label_encoder_differentiate = LabelEncoder()
label_encoder_grade = LabelEncoder()
label_encoder_a_stage = LabelEncoder()
label_encoder_estrogene = LabelEncoder()
label_encoder_progesterone = LabelEncoder()
label_encoder_status = LabelEncoder()
```

```python
X.iloc[:,1] = label_encoder_race.fit_transform(X.iloc[:,1])
#print(X.iloc[:,1])
X.iloc[:,2] = label_encoder_marital_status.fit_transform(X.iloc[:,2])
X.iloc[:,3] = label_encoder_t_stage.fit_transform(X.iloc[:,3])
X.iloc[:,4] = label_encoder_n_stage.fit_transform(X.iloc[:,4])
X.iloc[:,5] = label_encoder_6th_stage.fit_transform(X.iloc[:,5])
X.iloc[:,6] = label_encoder_differentiate.fit_transform(X.iloc[:,6])
X.iloc[:,7] = label_encoder_grade.fit_transform(X.iloc[:,7])
X.iloc[:,8] = label_encoder_a_stage.fit_transform(X.iloc[:,8])
X.iloc[:,10] = label_encoder_estrogene.fit_transform(X.iloc[:,10])
X.iloc[:,11] = label_encoder_progesterone.fit_transform(X.iloc[:,11])
```

```python
y = label_encoder_status.fit_transform(y)
```

### RandomOverSampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0) #creating a RandomOverSampler Object
```

```python
X_over, y_over = ros.fit_resample(X, y) #This function duplicates random samples of the minority value of the feature to increase it and be equal to the majority value
```

```python
sb.countplot(x=y_over) #This is a plot that count the occurance of a feature value
```

```python
#y_over.loc['Status'=='Alive']
y_over.valuecount()
```

### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(sampling_strategy='minority') #When str, specify the class targeted by the resampling. (it can be float, str, dict, or function)
#Here we specified only the minority class to be resampled
```

```python
X_smote, y_smote = smote.fit_resample(X,y) #fit_resample() with SMOTE only accepts numerical data
```

```python
sb.countplot(x=y_smote, tick_label=['Alive','Dead'])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search={'criterion':['gini','entropy'], 'max_depth':[100,200,None], 'min_samples_split':[2,10,20], 'min_samples_leaf':[1,2,10]}
dt=DecisionTreeClassifier()
dt_gs=GridSearchCV(estimator=dt, param_grid=grid_search, cv=4, verbose=-1, n_jobs=-1)
dt_gs.fit(X_train,y_train)
```

* cv: Determines the cross-validation splitting strategy.
* verbose: controls the output, the higher the verbose the more detailed the output
* n_jobs: number of jobs run by the processor in parallel, -1 means use all processors

```python
dt_gs.best_estimator_
```

```python
dt_gs.best_params_
```

```python
pd.DataFrame(dt_gs.cv_results_).sort_values(by='rank_test_score')
```
